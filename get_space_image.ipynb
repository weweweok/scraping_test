```python
#urlから目的の情報を抽出(賃貸の詳細情報の画像
import requests
import time
```

```python
from bs4 import BeautifulSoup
```

```python
def get_space_image(object_url):
  url = requests.get(object_url, timeout=5)
  soup = BeautifulSoup(url.text)
  result = soup.find_all('img',alt='間取り')

  def get_chitai_image(result):
    result_counter = 0 #imgデータの添え字
    for target in result:
      f = target.get('rel')
      result[result_counter] = f
      result_counter+=1
    return result
  result = get_chitai_image(result)
  url.close()
  return result
```

```python
object_url = 'https://suumo.jp/jj/chintai/ichiran/FR301FC005/?ar=030&bs=040&ta=13&sc=13114&sc=13120&sc=13116&sc=13201&cb=0.0&ct=3.0&mb=0&mt=9999999&et=9999999&cn=9999999&co=1&co=4&kz=1&shkr1=03&shkr2=03&shkr3=03&shkr4=03&sngz=&po1=25&po2=99&pc=100&page=1'
```

```python
img = get_space_image(object_url)
```
