```python
#urlから目的の情報を抽出(賃貸の名前
import requests
import time
```

```python
from bs4 import BeautifulSoup
```

```python
target_url = 'https://suumo.jp/jj/chintai/ichiran/FR301FC005/?ar=030&bs=040&ta=13&sc=13114&sc=13120&sc=13116&sc=13201&cb=0.0&ct=3.0&mb=0&mt=9999999&et=9999999&cn=9999999&co=1&co=4&kz=1&shkr1=03&shkr2=03&shkr3=03&shkr4=03&sngz=&po1=25&po2=99&pc=100'
```

```python
def get_name(target_url):
  for i in range(5):
    url = requests.get(target_url,timeout=5)#urlの取得
    soup = BeautifulSoup(url.text)#htmlの成型
    result = soup.find_all('a',class_='js-cassetLinkHref')#htmlからすべてのdiv要素をリストに格納(タグが邪魔になる)

  def name_wash(result):
    counter_cprice = 0
    for i in result:
      f = i.get_text()#tdタグ全体からテキストだけを取り出す
      result[counter_cprice] = f #参照したコンテナの上書き
      counter_cprice+=1            #次の配列を参照する添え字の準備

    for i in range(counter_cprice):
      result[i] = result[i].rstrip('\n')
    return result
  name = name_wash(result)
  url.close()
  return name
```

```python
name = get_name(target_url)
```

```python
for target in name:
  print(target)
```
